```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Dados do problema
m = 5.0  # Massa em kg
k = 100.0  # Constante da mola em N/m
u = 0.5  # Coeficiente de atrito de Coulomb
x0 = 0.4  # Posição inicial em m
v0 = 0  # Velocidade inicial em m/s
```

```python
# Definir a equação diferencial
def eq_diff(x, t):
    return [x[1], -(k/m)*x[0] - (u/m)*np.sign(x[1])]
```

```python
# Calcular a solução numérica da equação diferencial
t = np.linspace(0, 10, 1000)  # Vetor de tempo de 0 a 10 segundos
sol = np.array([x0, v0])  # Condições iniciais
x = np.zeros_like(t)  # Vetor de posição
for i in range(len(t)):
    x[i] = sol[0]
    sol = sol + 0.01*np.array(eq_diff(sol, t[i]))
```

```python
# Plotar o gráfico da resposta de vibração livre
plt.plot(t, x)
plt.xlabel('Tempo (s)')
plt.ylabel('Posição (m)')
plt.title('Resposta de vibração livre do sistema')
plt.show()
```
